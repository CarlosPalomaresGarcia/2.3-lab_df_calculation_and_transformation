```python
# Sub Problem 1: How to aggregate the  `amount_spent` for unique customers?

gastos_clientes = orders.groupby('CustomerID')['amount_spent'].sum()
# Le hago un groupby por el cliente con la suma de lo que se ha gastado cada cliente

# print the resulting series
print(gastos_clientes)
```

```python
#Sub Problem 2: How to select customers whose aggregated `amount_spent` is in a given quantile range?
# Identifico los VIP Customers

percentil_95 = orders['amount_spent'].quantile(0.95) # Para calcular el percentil 95

vip_customers = orders[orders['amount_spent'] <= percentil_95] # Selecciona las clientes que están dentro del percentil 95

# Identifico los Preferred Customers - entre el percentil 75 y 95.
percentil_75 = orders['amount_spent'].quantile(0.75) # Para calcular el percentil 75

preferred_customers = orders[(orders['amount_spent'] >= percentil_75) & (orders['amount_spent'] > percentil_95)]# Clientes que están entre el percentil 75 y 95

```

```python
display(vip_customers)
```

```python
display(preferred_customers)
```

```python
#Sub Problem 3: How to label selected customers as "VIP" or "Preferred"?
```

```python
#Pseudocodigo: Tengo que añadir una nueva columna en el que este el tipo de cliente
# Solo hay dos opciones, o VIP o prefered

# Asigno la palabra VIP a los clientes que esten en el percentil 95
orders['tipo_cliente'] = np.where(orders['amount_spent'] <= percentil_95, 'VIP', 'preferred')

#Añado los preferred que esten entre el 75 y 95 
orders.loc[(orders['amount_spent'] >= percentil_75) & (orders['amount_spent'] <= percentil_95), 'tipo_cliente'] = 'preferred'

# Mostrar el data frame actualizado
display(orders)
```
